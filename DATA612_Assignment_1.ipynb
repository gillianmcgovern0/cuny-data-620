## Import packages

```python
import networkx as nx
import matplotlib.pyplot as plot
```

## Create the kite graph

I decided to try to recreate the graph seen in the assignment 1 PDF. To do this, I used `networkx`'s `krackhardt_kite_graph` function:

```python
# Create the kite graph
g = nx.generators.krackhardt_kite_graph()
```

Since the graph in the PDF is the function's default graph, I did not add or remove any nodes. I did add the labels associated with each node:

```python
# add labels: Andre=1, Beverley=2, Carol=3, Diane=4, Ed=5, Fernando=6, Garth=7, Heather=8, Ike=9, Jane=10
labels = {0: 'Andre', 1: 'Beverley', 2: 'Carol', 3: 'Diane', 4: 'Ed', 5:'Fernando', 6:'Garth', 7:'Heather', 8:'Ike', 9:'Jane'}

# Plot the graph with the above labels
# Updated the node shape and color
nx.draw(g, node_shape="s", labels=labels, with_labels=True, node_color="red")
plot.draw()
```

We are now left with a graph looking similar to the graph shown in the assignment description.
